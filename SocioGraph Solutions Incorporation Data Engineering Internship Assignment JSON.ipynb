```python
# Read both files to a dataframe using pandas
df1 = pd.read_csv('customer_data_one.csv')
```

```python
df2 = pd.read_csv('customer_data_two.csv')
```

```python
# concatenate both the dataframes and ignore the indexes.
frames = [df1, df2]
result = pd.concat(frames, ignore_index = True)
```

```python
# dropping the duplicate columns and keeping the duplicated columns in the second dataset(new data)
result.drop_duplicates(subset = ['mobile_number'], keep = 'last' , inplace = True)
result
```

```python
# sorting based on first_names
result = result.set_index('first_name')
result = result.sort_index()
result = result.reset_index()
result
```

```python
# Converting Dataframe to JSON array of JSON objects.
# First Method: Use json library to convert dataframe to JSON, use orient = records

import json
res = json.loads(result.to_json(orient='records'))
print(res)
```

```python
# Second Method: use to_dict method to convert dataframe to dictionary.
res2 = result.to_dict(orient='records')
print(res2)
```

```python
#Checking if results of both methods are equal in type and content.
res == res2
```

```python
# writing the output to a file.
out = result.to_json(orient='records')[1:-1].replace('},{', '} {')
```

```python
with open('file_name.txt', 'w') as f:
    f.write(out)
```
